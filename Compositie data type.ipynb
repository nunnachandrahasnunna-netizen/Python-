## byte data types

```python
h=[20,30,50,60,70]
type(h)
```

```python
x= bytes(h)
```

```python
type(x)
```

```python
# # bytes data types is immutable
# x[0]=80
```

```python
for i in x:
    print(i)
```

```python
x[3]
```

## bytesarray data type

```python
y=[23,30,50,20]
```

```python
ba=bytearray(y)
```

```python
for i in y:
    print(i)
```

```python
ba[1]
```

```python
ba[0]=100
```

```python
for i in ba:
    print(i)
```
