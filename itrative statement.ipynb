## while loop

```python
x=1
while x<=10:
    print(x)
    x=x+1
```

```python
n=int(input("enter a num"))
sum=0
i=1
while i<=n:
    sum=sum+i
    i+=1
    
print("the sum of first",n,"number is",sum)
```

```python
cart=[100,220,250,300,350,700,900]
for iteam in cart:
    if iteam>500:
        print("you get 15 perrcent discount on this order")
        break
    print(iteam)
```

## continue

```python
for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
cart=[200,400,500,600,700]
for item in cart:
    if item>=500:
        print("we can not process this item",item)
        continue
    print(item)
```
